```python
#  fizzbuzz with list comprehension

num_list =[i for i in range(1,100+1)]

fizzbuzz_list =[]

for _item in num_list:
    if _item%15==0:
        _result ="fizzbuzz"
    else:
        if _item %3 ==0:
            _result ="fizz"
        else:
            if _item%5 ==0:
                _result ="buzz"
            else:
                _result = _item
    
    fizzbuzz_list.append(_result)
```

```python
fizzbuzz_list = ["fizzbuzz" if _item%15==0 \
                 else "fizz" if _item %3 ==0 \
                 else "buzz"if _item%5 ==0 \
                 else _item \
                 for _item in num_list]
fizzbuzz_list
```
