```python
def calculator(oper,*data):
    d_num=list(data)
    cnt=0
    sum1=0
    del1=0
    first=0
    
    if oper=="+":
        for _i in data:
            d_num[cnt]=_i
            print(d_num[cnt])
            sum1+=d_num[cnt]
            cnt+=1
        return sum1    
    elif oper=="-":
            for _i in data:
                d_num[cnt]=_i
                first=d_num[0]
                if cnt>=1:
                    del1=first-d_num[cnt]
                cnt+=1
            return del1    
    else:    
         print("Wrong Input")
            
calculator("+",1,2)
```
